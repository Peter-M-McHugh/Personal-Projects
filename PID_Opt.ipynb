```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl
import matplotlib.pyplot as plt

m = 1
b = 3
k = 6
# Define the transfer function in the Laplace domain
numerator = [1]
denominator = [m, b, k, 0]  # Example: s^2 + 2s + 1
sys = ctrl.TransferFunction(numerator, denominator)

# Time vector
time = np.linspace(0, 10, 1000)
step_input = np.ones_like(time)

# PID controller parameters
Kp = 4.5  # Proportional gain
Ki = 0  # Integral gain
Kd = 1.5  # Derivative gain

# PID transfer function
pid_numerator = [Kd, Kp, Ki]
pid_denominator = [1, 0]  # s term in the denominator for the integral part
pid_sys = ctrl.TransferFunction(pid_numerator, pid_denominator)

# Closed-loop system (plant with PID controller)
closed_loop_sys = ctrl.feedback(sys * pid_sys)
# Generate the step response of the system
#time, response = ctrl.step_response(sys, time)
response = ctrl.forced_response(closed_loop_sys, time, step_input)[1]

# Plot the time-domain response
plt.plot(time, response)
plt.plot(time,step_input)
plt.xlabel('Time')
plt.ylabel('System Response')
plt.title('Time-Domain Response of the Transfer Function')
plt.grid(True)
plt.show()
```
